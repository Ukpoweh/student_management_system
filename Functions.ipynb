## User-defined functions

```python
#define a function without parameters

def greet_user():
    print("Hello everyone")
```

```python
greet_user()
```

```python
#define a function with parameters

def greet_user_personalized(username):
    print(f"Hello {username}")
```

```python
greet_user_personalized("John")
```

```python
def greet_user_personalized():
    username = input("Please enter your name: ") #local variable
    print(f"Hello {username}")
```

```python
greet_user_personalized()
```

```python
#function with multiple parameters

def describe_student(name, age, course):
    print(f"My name is {name}. I am {age} years old. I am learning {course}")
```

```python
describe_student("Alice", 19, "Python") #positional arguments
```

```python
describe_student(19, "Alice", "Python")
```

```python
#keyword arguments

describe_student(course = "Python", age=19, name="Alice")
```

```python
#default values

def describe_student(name, age, course="Python"):
    print(f"My name is {name}. I am {age} years old. I am learning {course}")
```

```python
describe_student("Alice", 20)
```

```python
describe_student("Alice", 20, "SQL")
```

Variable scope

Local vs Global variable

Any variable defined within a function is a local variable and can't be accessed outside the function

```python
result = 8 #global variable
def sum_two_numbers(a, b):
    result = a + b #local variable
    return result
```

```python
print(result)
```

```python
first_number = 8
def difference(x, y):
    result = x-y
    return result
```

```python
print(first_number)
```

Passing an arbitrary number of arguments



```python
#function to add up numbers

def sum_of_numbers(*numbers):
    result = 0
    for num in numbers:
        result += num
    return result
```

```python
sum_of_numbers(1, 2, 3)
```

```python
sum_of_numbers(1, 2, 3, 4, 5, 20, 50, 67, 345, 78)
```

Anonymous/Lambda functions

```python
#to add two variables

add = lambda x,y: x + y 
```

```python
print(add(2, 3))
```

```python
print(add(2, 9))
```

```python
#using with the filter function

numbers = [1, 2, 3, 4, 5]
```

```python
#to filter the list to just even numbers

even_numbers = filter(lambda x: x % 2 == 0, numbers)
```

```python
print(list(even_numbers))
```

```python
#using with the map function

print(numbers)
```

```python
#to find the squares of all elements in the list

squares = map(lambda x: x ** 2, numbers)
```

```python
print(list(squares))
```
